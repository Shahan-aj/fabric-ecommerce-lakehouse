```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
%%sql
create table if not EXISTS Gold_Date
(
BusinessDate Date,
Business_Year int,
Business_Month int,
Business_Quarter int,
Business_Week int
)
using DELTA
PARTITIONED by (Business_Year)
```

```python
start_date='2015-01-01'
end_date='2030-12-31'

date_diff=spark.sql("select date_diff('{}','{}')".format(end_date,start_date)).first()[0]
```

```python
date_diff
```

now we will generate a range of numbers from 0 to 5843, use this to generate dates (adding these numbers to start date)

```python
ID_Data=spark.range(0,date_diff)
```

```python
ID_Data.show()
```

```python
date_data = ID_Data.selectExpr("date_add('{}',cast(id as int)) as BusinessDate".format(start_date))
```

```python
date_data.show()
```

```python
date_data.createOrReplaceTempView("ViewDate")
```

```python
%%sql
delete from Gold_Date;
```

```python
%%sql
insert into Gold_Date
select *, 
Year(BusinessDate) as Business_Year,
Month(BusinessDate) as Business_Month ,
Quarter(BusinessDate) as Business_Quarter ,
weekday(BusinessDate) as Business_Week  
from ViewDate
```
