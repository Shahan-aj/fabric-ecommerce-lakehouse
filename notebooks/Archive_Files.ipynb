```python
## now we need to archive the files in current folder, for that we create a notebook and mount to our files folder and move files to current folder to archive folder
## Microsoft fabric has provided an API to mount lakhouse to notebook

from notebookutils import mssparkutils
```

```python
mssparkutils.fs.help() # it gives a list of commands we can use when we're mounted
```

```python
mssparkutils.fs.mount("abfss://PCEU_Report_PRD@onelake.dfs.fabric.microsoft.com/Project_Lakehouse.Lakehouse/Files","/Files")
```

```python
mssparkutils.fs.getMountPath("/Files") # gives mount path for mount point
```

```python
check_files=mssparkutils.fs.ls(f"file://{mssparkutils.fs.getMountPath('/Files')}/Current/") # checking what all we have insid ecurrent
```

```python
check_files # using thsi variable we move one file at a time
```

```python
for file in check_files:
    if file.path:
        mssparkutils.fs.mv(file.path,f"file://{mssparkutils.fs.getMountPath('/Files')}/Archive/") 
```
